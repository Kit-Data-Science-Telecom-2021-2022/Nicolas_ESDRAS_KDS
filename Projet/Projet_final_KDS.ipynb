```python
url_glossaire = 'https://www.vendeeglobe.org/fr/glossaire'
requests.get(url_glossaire)
soup_glossaire = BeautifulSoup(requests.get(url_glossaire).content)

skipper_items = soup_glossaire.find_all('div', {'class':"sv-u-1 sv-u-s-1-2 sv-u-m-1-3 sv-u-l-1-6"})

# Utilisation d'un dictionnaire avant d'écrire le Dataframe.
boats_dict = {}

for soup_boat in skipper_items:
    boats_dict[soup_boat.find('h3', {'class':"boats-list__boat-name"}).text] = {
        'Skipper': soup_boat.find('span', {'class':"boats-list__skipper-name"}).text
    }

    for soup_li in soup_boat.find_all('li'):
        boats_dict[soup_boat.find('h3', {'class':"boats-list__boat-name"}).text][soup_li.text.split(':')[0]] = soup_li.text.split(':')[1]

df_boats = pd.DataFrame(boats_dict)
df_boats = df_boats.T
df_boats.dropna()
df_boats
```

De manière Visuel nous essayerons d'observer s'il y a une corrélation entre le rang et différentes variables

```python
VMG_rang_24h = pd.DataFrame(df_class.groupby('Rang')['VMG24 (kts)'].mean())
Rang = VMG_rang_24h.index.values.reshape(-1,1)
VMG_24h = VMG_rang_24h['VMG24 (kts)']

reg = LinearRegression(fit_intercept=True)
reg.fit(Rang, VMG_24h)
VMG_pred = reg.predict(Rang)

coef = np.round(reg.coef_[0],4)
intercept = np.round(reg.intercept_,4)
r2 = np.round(sklearn.metrics.r2_score(VMG_24h, VMG_pred),4)

print(f'Correlation entre le rang et la vitesse VMG depuis 24 heures:\n\n\
- Coefficient : {coef}\n\
- Intercept : {intercept}\n\
- Score R2 : {r2}\n')
```

Le Score R2 nous permet de déterminer s'il y a une forte corrélation en la rang du skipper et la vitesse VMG. Les résultats nous montrent que la corrélation est plutôt faible avec un R2 = 0,51
